```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x = np.arange(11)
x
```

```python
nfx = fuzz.trimf(x,[0,5,10])
nfx
```

```python
plt.plot(x,nfx)
```

```python
x2 = np.arange(10000)
```

```python
nfx2 = fuzz.trimf(x2,[min(x2),max(x2)/2,max(x2)])
nfx2
```

```python
plt.plot(x2,nfx2)
```
