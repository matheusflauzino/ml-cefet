```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_water = np.arange(0,20.5,0.5)
x_acidity = np.arange(0,12.5,0.5)
x_class = np.arange(0,1.05,0.5)
```

```python
print(list(x_water))
print(list(x_acidity))
print(list(x_class))
```

```python
#water content
water_lo = fuzz.trapmf(x_water,[0,0,2.5,10])
water_md = fuzz.trimf(x_water,[2.5,10,17.5])
water_hi = fuzz.trapmf(x_water,[10,17.5,20,20])

#index acidity
acidity_lo = fuzz.trapmf(x_acidity,[0,0,1.5,6])
acidity_md = fuzz.trapmf(x_acidity,[0,4.5,7.5,12])
acidity_hi = fuzz.trapmf(x_acidity,[6,10.5,12,12])

#class
class_a = fuzz.trapmf(x_class,[0,0,0.15,0.5])
class_b = fuzz.trimf(x_class,[0,0.5,1])
class_c = fuzz.trapmf(x_class,[0.5,0.85,1,1])
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))

ax0.plot(x_water, water_lo, 'b', linewidth=1.5, label='Low')
ax0.plot(x_water, water_md, 'g', linewidth=1.5, label='Mean')
ax0.plot(x_water, water_hi, 'r', linewidth=1.5, label='High')
ax0.set_title('Water Content')
ax0.legend()

ax1.plot(x_acidity, acidity_lo, 'b', linewidth=1.5, label='Small')
ax1.plot(x_acidity, acidity_md, 'g', linewidth=1.5, label='Mean')
ax1.plot(x_acidity, acidity_hi, 'r', linewidth=1.5, label='High')
ax1.set_title('Acitity Index')
ax1.legend()

ax2.plot(x_acidity, acidity_lo, 'b', linewidth=1.5, label='A')
ax2.plot(x_acidity, acidity_md, 'g', linewidth=1.5, label='Mean')
ax2.plot(x_acidity, acidity_hi, 'r', linewidth=1.5, label='High')
ax2.set_title('Acitity Index')
ax2.legend()
```
